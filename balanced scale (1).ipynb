```python
#Importing Libraries 
#basics and Visualization
import numpy as np 
import pandas as pd 

#Importing Libraries 
#basics and Visualization
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

#ML libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split


#metrics
from statistics import mean
from sklearn.metrics import accuracy_score as score
from sklearn.metrics import explained_variance_score as evs

#Ignore Warning 
import warnings as wrn
wrn.filterwarnings('ignore')
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/balance-scale/balance-scale.data"

# Assign colum names to the dataset
names = ['Class', 'L-Weight', 'L-Distance', 'R-Weight', 'R-Distance']

# Read dataset to pandas dataframe
dataset = pd.read_csv(url, names=names)
dataset.head()
```

```python
dataset['Class'] = LabelEncoder().fit_transform(dataset['Class'].tolist())
#pairplot
sns.pairplot(data=dataset)

#Heatmap
num_feat = dataset.select_dtypes(include=np.number).columns
plt.figure(figsize= (15, 15))
sns.heatmap(dataset.corr())
```

```python
y = dataset[['Class']]
X = dataset.drop(['Class'], axis = 1)

print(y.head())
print(X.head())
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 20)
```

```python
#Classification and prediction
#SVM

clf = SVC()
clf.fit(train_X, train_y)
pred = clf.predict(test_X)
print('Accuracy in percent = ',score(pred, test_y)*100)
```

```python
#Classification and prediction
#DT

clf = DecisionTreeClassifier()
clf.fit(train_X, train_y)
pred = clf.predict(test_X)
print('Accuracy in percent = ',score(pred, test_y)*100)
```
